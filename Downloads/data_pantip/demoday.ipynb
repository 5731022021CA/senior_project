```python
import pandas as pd
filter_pantip =[]
for l in range(3159,3280) :
    filename = "data-" + str(l)+".json"
    #print(filenam_jone)
    #print(type(filename))
    data = pd.read_json(filename,lines=True)
    for l in range(0,len(data)):
        sinthone = "sinthorn"
        if sinthone in data['forums'][l] :
            filter_pantip.append({'comments':data['comments'][l], 'forums':data['forums'][l],'id':data['id'][l],'message':data['message'][l],'timestamp':data['timestamp'][l],'title':data['title'][l]})
```

```python
filter_pantipdf = pd.DataFrame(filter_pantip)
filter_pantipdf
```

```python
data = []
for l in range(0,len(filter_pantipdf)) :  
    title_year = filter_pantipdf['timestamp'][l].year
    title_month = filter_pantipdf['timestamp'][l].month
    title_day = filter_pantipdf['timestamp'][l].day
    data.append({'ID':filter_pantipdf['id'][l],'text' : filter_pantipdf['title'][l], 'Timestamp':filter_pantipdf['timestamp'][l],'flag' : 'title'})
    data.append({'ID':filter_pantipdf['id'][l],'text' : filter_pantipdf['message'][l], 'Timestamp':filter_pantipdf['timestamp'][l],'flag' : 'message'}) 
    for r in filter_pantipdf['comments'][l]:
        comment_year = pd.to_datetime(r['timestamp'],unit='ms').year
        comment_month = pd.to_datetime(r['timestamp'],unit='ms').month
        comment_day = pd.to_datetime(r['timestamp'],unit='ms').day
        if title_year == comment_year and title_month == comment_month and title_day == comment_day :
            data.append({'ID':r['id'],'text':r['message'], 'Timestamp':pd.to_datetime(r['timestamp'],unit='ms'),'flag' : 'comment'}) 
        for m in r['replies'] :
            reply_year = pd.to_datetime(m['timestamp'],unit='ms').year
            reply_month = pd.to_datetime(m['timestamp'],unit='ms').month
            reply_day = pd.to_datetime(m['timestamp'],unit='ms').day
            if reply_year == reply_year and reply_month == comment_month and reply_day == comment_day :
                data.append({'ID':m['id'],'text':m['message'], 'Timestamp':pd.to_datetime(m['timestamp'],unit='ms'),'flag' : 'reply'}) 
```
